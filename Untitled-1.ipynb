```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
rest_data = pd.read_csv('datasets/rest_data_us_upd.csv',dtype={'object_type':'category','chain':'category'})
```

```python
rest_data.sample()
```

```python
rest_data.info()
```

```python
#Pie chart data

to_pie_type_groupby = rest_data.groupby(['object_type']).agg({'id':'count'}).reset_index()
to_pie_type_groupby.columns = ['object_type','count']

object_type = to_pie_type_groupby['object_type']
count = to_pie_type_groupby['count']
```

```python
#Pie chart

sns.set_palette('colorblind')
plt.figure(figsize=(5, 5))
plt.pie(count, labels=object_type,autopct='%1.1f%%',startangle=160)
plt.legend(object_type, loc="center left",bbox_to_anchor=(1, 0.5))
plt.title('Análise de tipos de estabelecimentos')

plt.show()
```

```python
#Pie chart data

to_pie_chain_groupby = rest_data.groupby(['chain']).agg({'id':'count'}).reset_index()
to_pie_chain_groupby.columns = ['chain','count']

chain = to_pie_chain_groupby['chain']
count_chain = to_pie_chain_groupby['count']
chain_legend = ['Not a chain','Chain']
```

```python
#Pie chart

sns.set_palette('colorblind')
plt.figure(figsize=(5, 5))
plt.pie(count_chain, labels=chain_legend,autopct='%1.1f%%',startangle=160)
plt.legend(chain_legend, loc="center left",bbox_to_anchor=(1, 0.5))
plt.title('Comparação entre restaurantes que são redes')
plt.show()
```

```python
chain_data = rest_data[rest_data['chain'] == 'True']
```

```python
# Countplot restaurantes de rede
sns.set_palette('colorblind')
plt.figure(figsize=(5, 5))
sns.countplot(data = chain_data, x = 'object_type')
plt.title('Comparação entre os tipos de restaurante que são redes')
plt.show()
print('O estabelecimento típico para redes são os restaurantes.\nO que vai de desencontro ao meu pensamento popular de que seria fast_food')
```

```python
# Countplot restaurantes de rede
plt.figure(figsize=(5, 5))
sns.histplot(data=rest_data, x='number',hue = 'chain', alpha = 0.5)
plt.legend(chain_legend, loc="center left",bbox_to_anchor=(1, 0.5))
plt.title('Frequência de número de assentos nos estabelecimentos')
plt.show()
```

Em comparação, os estabelecimentos de rede e os que não pertencem á rede tem uma frequência de até 50 assentos disponíveis por local. Quantidade mais alta no gráfico para os de rede pode ter conexão direta por essa categoria ser mais presente no nosso dataset, representando quase 62% de todos os estabelecimentos.

Nos dois gráficos acima, podemos observar que é mais comum que estabelecimentos de rede tenham uma quantidade superior do que estabelecimentos não identificados como de rede. Sendo assim, estabelecimentos de rede são caracterizados por terem muitos estabelecimentos com um pequeno número de assentos.

```python
seat_mean_type = pd.DataFrame(rest_data.groupby(['object_type']).agg({'number':'mean'}).reset_index())
seat_mean_type.columns = ['object_type','mean']
```

```python
sns.barplot(data=seat_mean_type, x='object_type', y='mean')
plt.title('Média de assentos nos diferentes tipos de estabelecimentos')
plt.xlabel('Tipo de estabelecimento')
plt.ylabel('Média')

plt.show()
```

```python
rest_data['street_name'] = rest_data['address'].str.replace(r'^\d+', '',regex = True).str.strip()
```

Construa um gráfico de dez ruas com o maior número de restaurantes.


```python
restaurant_data = rest_data[rest_data['object_type']=='Restaurant']
rest_per_street = restaurant_data.groupby(['street_name']).agg({'object_type':'count'}).reset_index().sort_values(by = 'object_type', ascending = False)
rest_per_street.columns = ['street_name','count']

ten_most_pop_street = rest_per_street.head(10)
```

```python
sns.barplot(data=ten_most_pop_street, x='street_name', y='count')
plt.title('Quantidades de restaurantes nas 10 ruas mais populares')
plt.xlabel('Rua')
plt.ylabel('Total')
plt.xticks(rotation = 40, ha='right')

plt.show()
```

```python
street_one_rest_ct = rest_per_street[rest_per_street['count']==1]['count'].count()
print(f'Total de ruas com apenas 1 restaurante: {street_one_rest_ct}')
```

```python
top_10_rest_data = rest_data.merge(ten_most_pop_street, on = 'street_name',how = 'inner')
sns.histplot(data=top_10_rest_data, x='number')
plt.title('Frequência de número de assentos para os restaurantes nas ruas mais populares')
plt.xlabel('Número de assentos')
plt.ylabel('Frequência')
plt.show()
```

```python
sns.boxplot(data=top_10_rest_data, x='number')
plt.title('Boxplot de número de assentos para os restaurantes nas ruas mais populares')
plt.xlabel('Número de assentos')
plt.show()
```
